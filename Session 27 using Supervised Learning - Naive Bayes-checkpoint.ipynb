```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("nba_2013.csv")
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
df.head()
```

```python
df.pts.head() # Since the column is continous/independent  we will convert to discrete/dependent variable to carry the regression
```

```python
# create a Series of booleans in which roughly half are True
nums = df.pts
mask_large = nums > 300

# initially set Size to small, then change roughly half to be large
df['Score points'] = 'small' # new column
df.loc[mask_large, 'pts'] = 'large'
```

```python
# create a new Series called IsLarge
df['Score_pts'] = df.pts.map({'small':0, 'large':1})
df.head()
```

```python
df.Score_pts.fillna(0,inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gaussian Naive Bayes
from sklearn import datasets
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
```

```python
X=df[['g', 'gs', 'mp', 'fg', 'fga']]
Y=df.Score_pts
```

```python
model = GaussianNB()
```

```python
from sklearn.model_selection import train_test_split
# Split the data into Trainging and Testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
model.fit(X_train,Y_train)
```

```python
# Predicted outcomes
predicted = model.predict(X_test)

# Actual Expected Outvomes
expected = Y_test
```

```python
 metrics.accuracy_score(expected, predicted)
```
